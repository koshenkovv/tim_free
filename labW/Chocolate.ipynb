Зависимости рейтинга компании от процента содержания какао в
шоколадной плитке.

```python
import seaborn as sns
```

```python
sns.lineplot(data=df, x="Rating", y="Cocoa\nPercent")
```

```python
a=df['Rating']
b=df['Cocoa\nPercent']
covariance = np.cov(a, b)[0][1]
print(covariance)
```

```python
import statsmodels.formula.api as smf
x1  = df['Rating']
y = df['Cocoa\nPercent']
model = smf.ols('y ~ x1', data=df)
res = model.fit()
print(res.summary())
```

Уравнение является статистически значимым, т.к. уровень
значимости критерия F (Prob.F-statistics = 2.12e-12%) для данного уравнения меньше принятого 5%

Получили параметры уравнения: a (свободный член уравнения) = 78,6437;
b (коэффициент полной регрессии) = -2,18, таким образом, полученное
уравнение имеет вид: y = 78,6437 -2,18x

Коэффициент полной регрессии показывает, что при изменении содержания какао в шоколаде на 1% рейтинг компании уменьшится на 2.18 пунктов

Коэффициент детерминации R2= 0.027 показывает, что 0,027% вариации рейтинга обусловлено процентным содержанием какао
